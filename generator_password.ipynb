```python
import random
import string

longueur = int(input("Longueur du mot de passe ? "))
if longueur < 8:
    print("Longueur minimum 8 — ajustée automatiquement à 8.")
    longueur = 8

chiffres = input("Des chiffres ? (oui/non) ")
symboles = input("Des symboles ? (oui/non) ")
nombre = int(input("Combien de mots de passe veux-tu générer ? "))

pool = string.ascii_letters
if chiffres == "oui":
    pool += string.digits
if symboles == "oui":
    pool += string.punctuation

for i in range(nombre):
    password = "".join(random.choices(pool, k=longueur))
    print(f"Mot de passe {i+1} : {password}")
```
