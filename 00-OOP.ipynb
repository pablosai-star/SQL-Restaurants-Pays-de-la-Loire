### Procedural code VS Object-Oriented Programming

<img src = "https://data-analytics-fullstack-assets.s3.eu-west-3.amazonaws.com/M05-Python_Programming/M2_D3_Procedural_Object.png"/>

Procedural code works fine when your project is small: a few functions, a couple of lists, and everything feels under control.
But as the system grows — more data, more rules, more interactions — your logic gets scattered, updates become fragile, and a simple change can break things in multiple places.

That's the moment where how you **organize** your code starts to matter as much as what it does.

### Our example

You want to manage a personal reading list. You track books: their title, author, number of pages, and how many times you've read them.
Your current approach uses separate functions for each task with disconnected data structures.

### The procedural approach breaks at scale

```python
# Procedural approach: separate data structures require manual coordination
book_titles = []
book_authors = []
book_pages = []
book_read_count = []

def add_book(title, author, pages):
    """Add a new book across four parallel lists"""
    book_titles.append(title)
    book_authors.append(author)
    book_pages.append(pages)
    book_read_count.append(0)
    return len(book_titles) - 1

def mark_as_read(book_index):
    """Update two separate lists when a book is read"""
    book_read_count[book_index] += 1
```

```python
# Add books and track indices manually
dune_idx = add_book("Dune", "Frank Herbert", 412)
hobbit_idx = add_book("The Hobbit", "J.R.R. Tolkien", 310)

mark_as_read(dune_idx)
mark_as_read(dune_idx)

print(f"{book_titles[dune_idx]}: read {book_read_count[dune_idx]} time(s), {book_pages[dune_idx]} pages")
```

This code demonstrates **three critical failures**:
- Manual index tracking creates fragility.
- Lose track of these numbers and you update the wrong book's data.
- Parallel list synchronization has no enforcement mechanism. Nothing prevents appending to `book_titles` without updating `book_authors`. Adding new attributes cascades changes everywhere. The approach cannot scale.

**Object-Oriented Programming** solves this by bundling related data with the functions that operate on that data. Instead of coordinating separate lists, you create `Book` objects that contain all their own information and know how to manage themselves.

## Classes and objects

A class defines **structure** and **behavior** as a blueprint.

An object creates an actual **instance** from that blueprint.

<img src = "https://data-analytics-fullstack-assets.s3.eu-west-3.amazonaws.com/M05-Python_Programming/M2_D3_Class_object.png"/>

When you write `class Book`, you define what all books share. When you write `dune = Book("Dune", "Frank Herbert", 412)`, you create an actual book with its own data.

Now let's take a look at the code.

```python
class Book:
    """
    Represents a book with reading tracking
    
    This is a CLASS DEFINITION - a blueprint for creating book objects.
    Think of it like a library card template: it describes what every book
    will have (attributes) and what every book can do (methods), but it
    doesn't create any actual books yet.
    """
    
    def __init__(self, title, author, pages):
        """
        Initialize a book with required parameters
        
        THE CONSTRUCTOR: This special method runs automatically when you
        create a new book object. Python calls it behind the scenes.
        
        Parameters:
            self: The object being created (Python passes this automatically)
            title: The book's title (required)
            author: The book's author (required)
            pages: Number of pages (required)
        
        When you write: dune = Book("Dune", "Frank Herbert", 412)
        Python actually does: Book.__init__(dune, "Dune", "Frank Herbert", 412)
        """
        # INSTANCE ATTRIBUTES: These belong to THIS specific book object
        # Each book gets their own separate copy of these values
        self.title = title          # Store the title on this book object
        self.author = author        # Store the author on this book object
        self.pages = pages          # Store the page count on this book object
        self.read_count = 0         # Initialize read counter to zero
        
        # The "self." prefix means "attach this to the object being created"
        # Without "self.", these would be local variables that disappear
        # when __init__ finishes running
    
    def mark_as_read(self):
        """
        Record that this book was read once more
        
        INSTANCE METHOD: A function that operates on a specific book object.
        
        Parameters:
            self: The book object calling this method (passed automatically)
        
        When you write: dune.mark_as_read()
        Python actually does: Book.mark_as_read(dune)
        """
        # Access THIS book's attributes using self
        # self.read_count refers to the read count for whichever book
        # called this method (could be dune.read_count, hobbit.read_count, etc.)
        self.read_count += 1
        return f'"{self.title}" marked as read (x{self.read_count})'
    
    def get_summary(self):
        """
        Generate a short book summary
        
        INSTANCE METHOD: Returns formatted information about this book.
        
        Parameters:
            self: The book object calling this method
        
        Returns:
            A string containing this book's key information
        """
        return f'"{self.title}" by {self.author} — {self.pages} pages, read {self.read_count} time(s)'
```

```python
# USAGE EXAMPLES to demonstrate how this works:

# Create two SEPARATE book objects from the same class blueprint
dune = Book("Dune", "Frank Herbert", 412)          # Creates one book object
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 310)  # Creates another book object

# Each book has their own independent attributes
# dune.title is "Dune", hobbit.title is "The Hobbit"
# They don't interfere with each other
```

```python
# Call instance methods on specific books
print(dune.mark_as_read())    # Updates DUNE's stats
print(dune.mark_as_read())    # Updates DUNE's stats again
print(hobbit.mark_as_read())  # Updates HOBBIT's stats (separate from Dune)
```

```python
# Each book maintains separate state
print(dune.get_summary())    # Shows: "Dune" by Frank Herbert — 412 pages, read 2 time(s)
print(hobbit.get_summary())  # Shows: "The Hobbit" by J.R.R. Tolkien — 310 pages, read 1 time(s)

# Key insight: dune and hobbit are DIFFERENT OBJECTS
# They both follow the Book blueprint, but they store different data
# Changing dune.read_count doesn't affect hobbit.read_count
```

```python
# Create two independent book objects from the same class
dune = Book("Dune", "Frank Herbert", 412)
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 310)

# Each book maintains completely separate state
print(dune.mark_as_read())
print(dune.mark_as_read())
print(hobbit.mark_as_read())
```

```python
# No index tracking needed
print(dune.get_summary())
print(hobbit.get_summary())
```

Each object owns its data completely. `dune` has been read 2 times while `hobbit` has been read 1 time. No manual index tracking appears. No parallel lists require synchronization. The book object bundles all its data together. Methods operate on that bundled data through the `self` parameter.

<Note type="important" title="Understanding self">

Every instance method receives `self` as its first parameter. Python automatically passes the object that called the method. When you write `dune.mark_as_read()`, Python translates this to `Book.mark_as_read(dune)`. The method receives `dune` as `self`, allowing it to access that specific object's attributes.

</Note>

## Attributes

**Instance attributes** belong to individual objects. When you write `self.read_count = 0` inside `__init__`, you create an attribute that belongs to that specific object. Each instance gets its own copy.

**Class attributes** are shared across all instances. When you write `minimum_pages = 1` directly in the class body, you create an attribute shared by all books.

```python
class Book:
    """Book with instance and class attributes"""
    
    # Class attributes: shared across all books
    minimum_pages = 1
    supported_formats = ["paperback", "hardcover", "ebook"]
    total_books = 0
    
    def __init__(self, title, author, pages):
        """Initialize instance attributes"""
        # Instance attributes: unique to this book
        self.title = title
        self.author = author
        self.pages = pages
        self.read_count = 0
        
        # Update class-level counter
        Book.total_books += 1
    
    def mark_as_read(self):
        self.read_count += 1
```

```python
# Create books - each has independent instance attributes
dune = Book("Dune", "Frank Herbert", 412)
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 310)
foundation = Book("Foundation", "Isaac Asimov", 255)
```

```python
dune.mark_as_read()
dune.mark_as_read()

print(f"Dune: read {dune.read_count} time(s)")
print(f"The Hobbit: read {hobbit.read_count} time(s)")
```

```python
# Class attributes are shared
print(f"\nMinimum pages allowed: {Book.minimum_pages}")
print(f"Total books created: {Book.total_books}")
```

## Encapsulation

Direct attribute access creates data corruption. When any code can write `book.read_count = -5` or `book.pages = "many"`, your program accepts nonsense. The book object cannot defend its own integrity.

Encapsulation transfers control from callers to the object itself. The object declares that external code cannot touch internal data directly. Every change must pass through methods or properties that enforce rules.

**Python uses the underscore prefix to signal private attributes.** Writing `_read_count` tells other developers this attribute belongs to the class implementation, not the public interface.

<Note type="important">

The underscore creates a convention, not a lock. Python trusts developers to respect boundaries rather than enforcing them through compiler errors.

</Note>

Properties provide the enforcement mechanism. When you write `book.pages`, Python routes that access through a property method. That method runs validation code before allowing the change. Invalid values get rejected. The syntax looks like simple attribute access, but the property intercepts every operation.

Consider this unprotected implementation:

```python
class Book:
    def __init__(self, title):
        self.title = title
        self.pages = 0
        self.read_count = 0
```

Any code can now corrupt the book's state:

```python
book = Book("Dune")
book.pages = -50          # Accepts impossible value
book.read_count = "many"  # Accepts wrong type
book.colour = "blue"      # Creates attributes that shouldn't exist
```

Anyone can modify any attribute with any value. No validation exists. Typos create new attributes silently. This breaks in production when negative page counts corrupt export logic or string read counts break statistics.

**Encapsulation solves this by hiding internal data and exposing only controlled methods:**

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self._read_count = 0      # Underscore signals "private - don't touch"
        self._pages = 0
    
    def mark_as_read(self):
        self._read_count += 1     # Only increments - never decrements
        return f'Read x{self._read_count}'
    
    @property
    def read_count(self):         # Read-only access through property
        return self._read_count
    
    @property
    def pages(self):
        return self._pages
```

```python
dune = Book("Dune", "Frank Herbert")
dune.mark_as_read()
print(dune.read_count)  # Output: 1

try:
    dune.read_count = -50
except AttributeError as e:
    print(f"Assignment blocked: {e}")
    # Output: Assignment blocked: can't set attribute
```

The `@property` decorator transforms the `read_count` method into a read-only attribute. External code can call `dune.read_count` to retrieve the value, but any attempt to assign `dune.read_count = -50` triggers an `AttributeError`. Python refuses the operation because no setter method exists.

The `mark_as_read` method provides the only valid path to change the read count. This method accesses `_read_count` directly because it lives inside the class. The method enforces business rules: the count only increments by one, it never decreases, and it never accepts arbitrary values from outside.

This pattern prevents three corruption scenarios. First, negative read counts become impossible because `mark_as_read` only adds positive increments. Second, type errors disappear because the method controls all assignments. Third, the read count always reflects actual reading sessions rather than random values injected from external code.

The object now owns its data integrity. External code cannot bypass validation rules or create inconsistent state. Every read count you retrieve accurately represents sessions that passed through `mark_as_read`.

<Note type="tip">

A setter lets you control writes to a property while still using assignment syntax. You add validation logic that runs before Python stores the value. When you write `book.pages = 412`, Python calls your setter method, which checks whether 412 passes validation rules before updating the private attribute.

Without a setter, the property becomes read-only. With a setter, you maintain the natural assignment syntax while enforcing constraints that protect data integrity.

</Note>

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self._read_count = 0
        self._pages = 0
    
    @property
    def pages(self):
        return self._pages
    
    @pages.setter
    def pages(self, value):
        if not isinstance(value, int):
            raise TypeError("Pages must be an integer")
        if value < 1:
            raise ValueError("A book must have at least one page")
        self._pages = value
```

The setter method runs every time external code assigns to `pages`. The validation checks type and range before updating `_pages`. Invalid values trigger exceptions that prevent corruption:

```python
dune = Book("Dune", "Frank Herbert")
dune.pages = 412   # Works - passes validation
print(dune.pages)
```

```python
try:
    dune.pages = "many"
except TypeError as e:
    print(f"Assignment blocked: {e}")
    # Output: Assignment blocked: Pages must be an integer
```

## Method types

In Python, methods come in **three main flavors** depending on what they act on: the individual object, the class itself, or neither (just a related utility function).

- Instance methods receive `self` and operate on **one specific object**. They can read and modify that object's state: `book.mark_as_read()` updates *this* book only.

- Class methods receive `cls` and operate on the **class as a whole**. They're often used for alternative constructors or for actions that concern all instances, e.g. `Book.from_dict(...)` or `Book.total_books`.

- Static methods receive neither `self` nor `cls`. They're **utility functions** that are related to the class conceptually, but don't need access to instance or class state, e.g. `Book.validate_isbn(isbn)`. They live inside the class to keep related logic grouped in one place.

```python
class Book:
    """Book with different method types"""
    
    total_books = 0
    book_registry = {}
    
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
        self.read_count = 0
        
        # Register book by title
        Book.total_books += 1
        Book.book_registry[title] = self
    
    # Instance method - operates on a specific book
    def mark_as_read(self):
        """Increment this book's read count"""
        self.read_count += 1
        return f'"{self.title}" read x{self.read_count}'
    
    # Class method - operates on class-level data
    @classmethod
    def get_book_count(cls):
        """Return total number of books created"""
        return cls.total_books
    
    # Class method - alternative constructor
    @classmethod
    def find_book(cls, title):
        """Look up a book by its title"""
        return cls.book_registry.get(title)
    
    # Static method - utility function
    @staticmethod
    def validate_isbn(isbn):
        """Check if an ISBN has a valid format (simplified: 13 digits)"""
        return isinstance(isbn, str) and isbn.isdigit() and len(isbn) == 13
```

```python
# Create books
dune = Book("Dune", "Frank Herbert", 826)
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 400)
```

```python
# Instance method works with a specific book
print(dune.mark_as_read())
```

```python
# Class method works with class-level data
print(f"\nTotal books: {Book.get_book_count()}")

# Class method looks up from registry
found = Book.find_book("Dune")
print(f"Found: {found.author if found else 'Not found'}")
```

```python
# Static method validates without instance or class state
print(f"\n'9782070360024' valid ISBN: {Book.validate_isbn('9782070360024')}")
print(f"'123' valid ISBN: {Book.validate_isbn('123')}")
```

## Special methods

Special methods integrate objects with built-in operations. Define `__str__` to control string conversion. Define `__eq__` to support equality comparison. Define `__lt__` to enable sorting.

| Method | Controls | Example Usage |
|--------|----------|---------------|
| `__str__` | String representation for humans | `print(book)` calls `book.__str__()` |
| `__repr__` | String representation for developers | Console display and debugging output |
| `__eq__` | Equality comparison | `book1 == book2` calls `book1.__eq__(book2)` |
| `__lt__` | Less-than comparison | `book1 < book2` enables sorting operations |
| `__len__` | Length calculation | `len(shelf)` returns number of books |
| `__add__` | Addition operator | `shelf1 + shelf2` merges two shelves |
| `__getitem__` | Index access | `shelf[0]` retrieves first book |
| `__iter__` | Iteration protocol | `for book in shelf` loops through books |

Python calls these methods automatically when your objects encounter built-in operations. You write the method once, and Python integrates your object with the language's standard syntax. This transforms custom objects into first-class citizens that behave like built-in types.

```python
class Book:
    """Book with special methods for natural operations"""
    
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
        self.read_count = 0
        self.rating = None   # None until rated
    
    def mark_as_read(self, rating=None):
        self.read_count += 1
        if rating is not None:
            self.rating = max(0, min(5, rating))  # clamp between 0 and 5
    
    def __str__(self):
        """User-friendly string representation"""
        stars = f" ({'★' * self.rating}{'☆' * (5 - self.rating)})" if self.rating else ""
        return f'"{self.title}" by {self.author}{stars}'
    
    def __repr__(self):
        """Developer-focused representation"""
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    def __eq__(self, other):
        """Two books are equal if they share the same title and author"""
        if not isinstance(other, Book):
            return False
        return self.title == other.title and self.author == other.author
    
    def __lt__(self, other):
        """Sort by rating first, then by page count"""
        r1 = self.rating or 0
        r2 = other.rating or 0
        if r1 != r2:
            return r1 < r2
        return self.pages < other.pages
    
    def __hash__(self):
        """Make books hashable for use in sets/dicts"""
        return hash((self.title, self.author))
```

```python
# Create and rate books
dune = Book("Dune", "Frank Herbert", 412)
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 310)
foundation = Book("Foundation", "Isaac Asimov", 255)

dune.mark_as_read(rating=5)
hobbit.mark_as_read(rating=4)
```

```python
# __str__ for user display
print("User view:")
print(f"  {dune}")
```

```python
# __repr__ for development
print(f"\nDeveloper view: {repr(hobbit)}")
```

```python
# __eq__ for equality
dune2 = Book("Dune", "Frank Herbert", 412)
print(f"\ndune == dune2: {dune == dune2}")
print(f"dune == hobbit: {dune == hobbit}")
```

```python
# __lt__ enables sorting
books = [dune, hobbit, foundation]
books.sort(reverse=True)
print("\nBooks ranked by rating:")
for i, book in enumerate(books, 1):
    print(f"  {i}. {book}")
```

## Complete book management system

This complete example integrates all fundamental OOP concepts into a working personal library.

```python
class Book:
    """Complete book management with all OOP fundamentals"""
    
    # Class attributes
    minimum_pages = 1
    total_books = 0
    book_registry = {}
    
    def __init__(self, title, author, pages):
        # Validation
        if not title or not author:
            raise ValueError("Title and author are required")
        if not isinstance(pages, int) or pages < self.minimum_pages:
            raise ValueError(f"Pages must be an integer >= {self.minimum_pages}")
        
        # Instance attributes
        self.title = title
        self.author = author
        self._pages = pages
        self._read_count = 0
        self._rating = None
        self.status = 'unread'
        
        # Register
        Book.total_books += 1
        Book.book_registry[title] = self
    
    # Properties for controlled access
    @property
    def pages(self):
        return self._pages
    
    @property
    def read_count(self):
        return self._read_count
    
    @property
    def rating(self):
        return self._rating
    
    @property
    def is_favourite(self):
        return self._rating is not None and self._rating >= 4
    
    @property
    def reading_time_hours(self):
        """Estimate reading time at ~250 words/page, ~200 words/min"""
        words = self._pages * 250
        return round(words / (200 * 60), 1)
    
    # Instance methods
    def mark_as_read(self, rating=None):
        self._read_count += 1
        self.status = 'read'
        if rating is not None:
            if not isinstance(rating, (int, float)) or not 0 <= rating <= 5:
                raise ValueError("Rating must be a number between 0 and 5")
            self._rating = rating
        return f'"{self.title}" marked as read (x{self._read_count})'
    
    def get_summary(self):
        stars = f"{'★' * int(self._rating)}" if self._rating else "not rated"
        fav = " ❤️" if self.is_favourite else ""
        return f"""
  "{self.title}" — {self.author}{fav}
  Pages: {self._pages} (~{self.reading_time_hours}h)
  Read: {self._read_count} time(s)
  Rating: {stars}
  Status: {self.status}
        """
    
    # Class methods
    @classmethod
    def get_book_count(cls):
        return cls.total_books
    
    @classmethod
    def find_book(cls, title):
        return cls.book_registry.get(title)
    
    @classmethod
    def list_all_books(cls):
        return list(cls.book_registry.keys())
    
    # Static method
    @staticmethod
    def validate_isbn(isbn):
        return isinstance(isbn, str) and isbn.isdigit() and len(isbn) == 13
    
    # Special methods
    def __str__(self):
        stars = f" ({'★' * int(self._rating)})" if self._rating else ""
        return f'"{self.title}" by {self.author}{stars}'
    
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self._pages})"
    
    def __eq__(self, other):
        if not isinstance(other, Book):
            return False
        return self.title == other.title and self.author == other.author
```

```python
# Build a reading list
dune = Book("Dune", "Frank Herbert", 412)
hobbit = Book("The Hobbit", "J.R.R. Tolkien", 310)
foundation = Book("Foundation", "Isaac Asimov", 255)
```

```python
# Mark books as read with ratings
print("=== READING SESSION ===")
print(dune.mark_as_read(rating=5))
print(dune.mark_as_read())          # Read again, no new rating
print(hobbit.mark_as_read(rating=4))
print(foundation.mark_as_read(rating=3))
```

```python
# Display summaries
print("\n=== MY LIBRARY ===")
print(dune.get_summary())
print(hobbit.get_summary())
```

```python
# Use class methods
print("=== LIBRARY STATS ===")
print(f"Total books: {Book.get_book_count()}")
print(f"All titles: {', '.join(Book.list_all_books())}")
```

```python
# Lookup a book
found = Book.find_book("Dune")
print(f"\nFound: {found}")
```

```python
# Validate an ISBN
print(f"\nISBN validation:")
print(f"  '9782070360024': {Book.validate_isbn('9782070360024')}")
print(f"  'not-an-isbn': {Book.validate_isbn('not-an-isbn')}")
```
