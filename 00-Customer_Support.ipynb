## Task 6: Calculate response time

Tickets must be answered within 4 hours. Sarah's ticket arrived at 09:15 (represented as 9.25 hours in decimal). Marcus responded at 11:45 (11.75 hours). Calculate the response time in hours and minutes.

<Note type="hint">
Use basic arithmetic operators: `+`, `-`, `*`, `/`. Use `//` for integer division and `%` for remainder. Round decimals with `round(number, decimals)`.
</Note>

````python
# Time values in decimal hours
ticket_time = 9.25  # 09:15 AM
response_time = 11.75  # 11:45 AM

# Calculate response time
time_difference = response_time - ticket_time

# Convert to hours and minutes
hours = int(time_difference)
minutes = int((time_difference - hours) * 60)

# Print formatted result
print(f"Response time: {time_difference} hours")
print(f"Response time: {hours} hours and {minutes} minutes")
````

## Task 16: Compare ticket priorities

TechFlow assigns numeric priority scores to tickets. Higher numbers mean higher priority. You need to compare tickets to determine which requires attention first.

Three tickets arrived:
- Ticket A: Priority score 85
- Ticket B: Priority score 92
- Ticket C: Priority score 78

Determine which ticket has the highest priority. Calculate the difference between the highest and lowest priority scores. Check if any ticket exceeds the urgent threshold of 90.

Store the results as boolean values and print descriptive messages.

````python
# Ticket priority scores
priority_a = 85
priority_b = 92
priority_c = 78
URGENT_THRESHOLD = 90

# Find highest priority
highest_priority = max(priority_a, priority_b, priority_c)
lowest_priority = min(priority_a, priority_b, priority_c)

# Calculate priority spread
priority_spread = highest_priority - lowest_priority

# Check urgent status
a_urgent = priority_a >= URGENT_THRESHOLD
b_urgent = priority_b >= URGENT_THRESHOLD
c_urgent = priority_c >= URGENT_THRESHOLD

# Print analysis
print(f"Highest priority: {highest_priority}")
print(f"Lowest priority: {lowest_priority}")
print(f"Priority spread: {priority_spread}")
print(f"Ticket A urgent: {a_urgent}")
print(f"Ticket B urgent: {b_urgent}")
print(f"Ticket C urgent: {c_urgent}")
````

## Task 20: Verify data types and conversions

The quality assurance team needs to verify that ticket data uses correct types before database insertion. Check the types of various ticket components and convert any that are incorrect.

You receive this mixed data:
- `ticket_id = "4521"` (should be integer)
- `customer_age = "32"` (should be integer)
- `satisfaction_score = 4` (should be float)
- `ticket_count = 145.0` (should be integer)
- `is_resolved = "True"` (should be boolean)

Check each variable's type using `isinstance()`. Convert each to the correct type. Verify the conversion worked by checking types again. Print before and after types for each variable.

<Note type="hint">
Use `isinstance(variable, type)` to check types. For boolean conversion from string, use comparison: `variable == "True"` rather than `bool()` which always returns True for non-empty strings.
</Note>

````python
# Mixed data types
ticket_id = "4521"
customer_age = "32"
satisfaction_score = 4
ticket_count = 145.0
is_resolved = "True"

# Check original types
print("=== Original Types ===")
print(f"ticket_id: {type(ticket_id)} - {isinstance(ticket_id, int)}")
print(f"customer_age: {type(customer_age).__name__} - {isinstance(customer_age, int)}")
print(f"satisfaction_score: {type(satisfaction_score).__name__} - {isinstance(satisfaction_score, float)}")
print(f"ticket_count: {type(ticket_count).__name__} - {isinstance(ticket_count, int)}")
print(f"is_resolved: {type(is_resolved).__name__} - {isinstance(is_resolved, bool)}")


# Convert to correct types
ticket_id = int(ticket_id)
customer_age = int(customer_age)
satisfaction_score = float(satisfaction_score)
ticket_count = int(ticket_count)
is_resolved = is_resolved == "True"

# Verify new types
print("\n=== Converted Types ===")
print(f"ticket_id: {type(ticket_id).__name__} - {isinstance(ticket_id, int)}")
print(f"customer_age: {type(customer_age).__name__} - {isinstance(customer_age, int)}")
print(f"satisfaction_score: {type(satisfaction_score).__name__} - {isinstance(satisfaction_score, float)}")
print(f"ticket_count: {type(ticket_count).__name__} - {isinstance(ticket_count, int)}")
print(f"is_resolved: {type(is_resolved).__name__} - {isinstance(is_resolved, bool)}")


# Print results
print("\n=== Values ===")
print(f"ticket_id: {ticket_id}")
print(f"customer_age: {customer_age}")
print(f"satisfaction_score: {satisfaction_score}")
print(f"ticket_count: {ticket_count}")
print(f"is_resolved: {is_resolved}")
````
