## Task 10: Monitor inventory reorder points

Products need reordering when stock falls below the reorder point. Check these products and their current stock levels:
- Product A: 150 units, reorder point 200
- Product B: 75 units, reorder point 50
- Product C: 25 units, reorder point 30
- Product D: 180 units, reorder point 100

Store the data in two lists: stock levels and reorder points. Use a `for` loop with `range()` to check each product by index. If stock is below the reorder point, print a reorder alert. Count how many products need reordering.

<Note type="hint">
Use `range(len(list))` to get indices. Access list items with `list[index]`. Compare values at the same index from both lists inside the loop.
</Note>

```python
# Product data
product_names = ["Product A", "Product B", "Product C", "Product D"]
stock_levels = [150, 75, 25, 180]
reorder_points = [200, 50, 30, 100]

# Check reorder requirements
reorder_needed = 0

for i in range(len(product_names)):
    if stock_levels[i] < reorder_points[i]:
        shortage = reorder_points[i] - stock_levels[i]
        print(f"REORDER ALERT: {product_names[i]} - Stock: {stock_levels[i]}, Reorder point: {reorder_points[i]}, Short by: {shortage}")
        reorder_needed += 1
    else:
        print(f"OK: {product_names[i]} - Stock: {stock_levels[i]}, Reorder point: {reorder_points[i]}")

print(f"\nProducts needing reorder: {reorder_needed}")
```

## Task 11: Calculate order processing time

ShopStream tracks how long it takes to process each order. The system should continue processing orders until total processing time exceeds 60 minutes per batch.

Each order takes a random amount of time. Start with 0 minutes. Add processing time for each order: 12 minutes, 8 minutes, 15 minutes, 22 minutes, 18 minutes, 25 minutes.

Use a `while` loop to process orders one by one. Track accumulated time. Stop when total time would exceed 60 minutes. Print each order's processing time and the final count of orders processed.

<Note type="hint">
Store processing times in a list. Keep a counter to track which order you are processing. Use the counter as an index. Check if adding the next order time would exceed the limit before processing it.
</Note>

```python
# Processing times for orders (in minutes)
processing_times = [12, 8, 15, 22, 18, 25]
time_limit = 60

# Process orders until time limit reached
total_time = 0
orders_processed = 0
order_index = 0

while order_index < len(processing_times):
    next_order_time = processing_times[order_index]

    if total_time + next_order_time > time_limit:
        print(f"Cannot process order {order_index + 1} ({next_order_time} min) - would exceed time limit")
        break

    total_time += next_order_time
    orders_processed += 1
    print(f"Processed order {orders_processed}: {next_order_time} minutes (total: {total_time} minutes)")
    order_index += 1

print(f"\nOrders processed: {orders_processed}")
print(f"Total time used: {total_time} minutes")
```
