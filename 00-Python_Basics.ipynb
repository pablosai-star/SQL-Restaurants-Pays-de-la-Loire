## Variables
One of the first things, that you have to learn is a **variable** where it is a named container that **stores data** (text, number, etc.) in computer memory. Once the variable is compute, its data can be access everywhere. The type of the variable is inferred based on the value assigned. It means that it **recognizes** the store data by itself.

````python
city_name = "Paris"
ZIP = "75001"
````

We have seen our first initialization or assignment. Based on the type of data that you used you can do many **manipulation** such as calculation, **rewriting** text or even **reassigning** it

````python
# Reassignment
city_name = "Marseille"
ZIP = "13001"
````

By doing so, the variable's data is **overwritten** with the new one and the previous data is **deleted**.

<Note type="note">

Knowing how to name your variables is important to make your code more readable. Here are some good practices:

- **Descriptive names**: ```total_residents``` instead of ```tr```
- **Snake_case**: ```average_income_per_resident``` not averageIncomeperResident
- **Avoid reserved words**: Don't use ```class```, ```def```, etc.
- **Constants**: Use ALL_CAPS for constants ```ZIP``` = 75003
- **Consistent naming**: Use the same naming convention for all variables in a function

### Convert a data type to another
We just created values of **specific types** (ints, floats, strings, …); next we’ll convert between **types** because in real data, values often arrive as text and must be **cast** before we can compute, compare, or sort correctly.

## Check Type 

When working with users or external data, it's important to **check the type** of your variables before using them. Python gives you two useful tools for this: `input()` and `isinstance()`.

### Get data from the user

The `input()` function lets you ask the user to type something.

```python
city = input("Enter your city: ")
print(city)
```

By default, `input()` always returns a **string**, even if the user types a number.

To use the value as a number, you must convert it:

```python
age = input("Enter your age: ")     # User types 30 → still a string
age = int(age)                      # Convert to integer
print(age + 1)                      # Now it works!
```

### Check the data type

If you're not sure what type a variable has, use `isinstance()`:

```python
print(isinstance(age, int))       # True
print(isinstance(city, str))      # True
```

You can also check against multiple types:

```python
print(isinstance(age, (int, float)))  # Is age a number?
```

This helps avoid errors before doing operations. Always check before converting or calculating. These two tools: `input()` for collecting data and `isinstance()` for checking it—are essential when your program starts interacting with real people and unpredictable inputs.
