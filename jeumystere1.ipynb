```python
import random

print("Choisis ton niveau :")
print("1 → Facile (1-50)")
print("2 → Normal (1-100)")
print("3 → Difficile (1-500)")

historique=[]
choix_niveau = int(input("Ton choix : "))

if choix_niveau == 1:
    limite = 50
    essais_max = 10
elif choix_niveau == 2:
    limite = 100
    essais_max = 7
else:
    limite = 500
    essais_max = 5

nombre_mystere = random.randint(1, limite)
compteur = 0

while True:
    essai = int(input(f"Saisis un nombre entre 1 et {limite} : "))
    historique.append(essai)  # ← la liste qui appelle append
    compteur += 1
    
    if compteur >= essais_max:
        print(f"Game Over ! Le nombre mystère était {nombre_mystere}")
        print(f"Tes essais : {historique}")  # ← avant le break
        break
    
    if essai > nombre_mystere:
        print("Trop grand !")
    elif essai < nombre_mystere:
        print("Trop petit !")
    else:
        print(f"Bravo ! Tu as trouvé en {compteur} essais !")
        print(f"Tes essais : {historique}")  # ← avant le break
        break
```

```python
import pandas as pd

transactions = []

while True:
    print("1 → Ajouter une transaction")
    print("2 → Afficher le solde")
    print("3 → Afficher le résumé")
    print("4 → Quitter")
    
    choix = int(input("Ton choix : "))
    
    if choix == 1:
        print("Tu as choisi : Ajouter une transaction")
        type_transaction = input("Dépense ou revenu ? ")
        montant = int(input("Quel est le montant ? "))
        categorie = input("Quelle catégorie ? ")
        
        transaction = {"type": type_transaction, "montant": montant, "categorie": categorie}
        transactions.append(transaction)
    elif choix == 2:
   
        solde = 0
    
        for transaction in transactions:
           if transaction["type"] == "revenu":
               solde += transaction["montant"]
           elif transaction["type"] == "dépense":
               solde -= transaction["montant"]
    
        print(f"Ton solde est de {solde} €")
    elif choix == 3:
        
        resume = {}
        
        for transaction in transactions:
            categorie = transaction["categorie"]
            
            if categorie not in resume:
                resume[categorie] = 0  # on initialise la catégorie à 0
            
            if transaction["type"] == "revenu":
                resume[categorie] += transaction["montant"]
            elif transaction["type"] == "dépense":
                resume[categorie] -= transaction["montant"]
        
        print("Résumé par catégorie :")
        for cle, valeur in resume.items():
            print(f"{cle} : {valeur} €")
    elif choix == 4:
        print("Au revoir !")
        break
    elif choix == 5:
        df = pd.DataFrame(transactions)
        df.to_csv(r"D:\DATA_FULL_STACK\Data\budget.csv", index=False)
        print("Transactions sauvegardées !")
```

```python
notes = [
    {"matiere": "Maths", "note": 12},
    {"matiere": "Python", "note": 15},
    {"matiere": "Maths", "note": 8}
]

total_par_matiere = {}

for note in notes:
    matiere = note["matiere"]
    if matiere not in total_par_matiere:
        total_par_matiere[matiere] = 0
    total_par_matiere[matiere] += note["note"]

print(total_par_matiere)
```

```python
ventes = [
    {"produit": "stylo", "quantite": 5, "operation": "achat"},
    {"produit": "stylo", "quantite": 2, "operation": "vente"},
    {"produit": "cahier", "quantite": 10, "operation": "achat"}
]

stock={}

for vente in ventes:
    produit = vente["produit"]  # ← vente, pas ventes
    
    if produit not in stock:
        stock[produit] = 0
    
    if vente["operation"] == "achat":
        stock[produit] += vente["quantite"]
    elif vente["operation"] == "vente":
        stock[produit] -= vente["quantite"]

# resultat attendu: stock = {"stylo": 3, "cahier": 10}
```

```python
import pandas as pd
import os

# Charger les contacts existants si le fichier existe
if os.path.exists(r"D:\DATA_FULL_STACK\Data\contacts.csv"):
    df = pd.read_csv(r"D:\DATA_FULL_STACK\Data\contacts.csv")
    contacts = df.to_dict("records")  # convertit le DataFrame en liste de dicts
else:
    contacts = []

while True:
    print("1 → Ajouter un contact")
    print("2 → Rechercher un contact")
    print("3 → Afficher tous les contacts")
    print("4 → Supprimer un contact")
    print("5 → Sauvegarder en CSV")
    print("6 → Quitter")

    try:
        choix = int(input("Ton choix : "))
    except ValueError:
        print("Saisis un chiffre entre 1 et 6 !")
        continue  # repart au début de la boucle

    if choix == 1:
        print("Tu as choisi : Ajouter un contact")
        #{"nom": "Giovanni", "telephone": "0612345678", "email": "giovanni@gmail.com"}
        prenom=input("Quel est ton prenom?")
        telephone=input("Quel est ton numero?")
        adr_mail=input("Quelle est ton adresse mail?")
        contact={"prenom":prenom, "telephone": telephone, "email":adr_mail}
        contacts.append(contact)
    elif choix == 2:
        print("Tu as choisi : Rechercher un contact")
    elif choix == 3:
        print("Tu as choisi : Afficher tous les contacts")
        #Prénom : Giovanni | Téléphone : 0612345678 | Email : giovanni@gmail.com
    
        for contact in contacts:
            print(f"Prénom : {contact['prenom']} | Téléphone : {contact['telephone']} | Email : {contact['email']}")    

    elif choix == 4:
        print("Tu as choisi : Supprimer un contact")
    elif choix == 5:
        print("Tu as choisi : Sauvegarder en CSV")
        df = pd.DataFrame(contacts)
        df.to_csv(r"D:\DATA_FULL_STACK\Data\contacts.csv", index=False)
        print("Contacts sauvegardées !")
    elif choix == 6:
        print("Au revoir !")
        break
```
